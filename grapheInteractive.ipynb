```python
# @title
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact_manual, FloatText
def plot_graph(a,b,c):

  def kostil_point_touche(a, b, c):
    x = np.sqrt(-1/(c*a))
    z = a*x+b-(1/(c*x))
    return(z)

  z = kostil_point_touche(a, b, c)

  def indifference(x):
      return 1/(c*x)+z

  def contrainteBudgetaire(x):
      return(a*x+b)
  # graphique
  #
  borderlimit = 15
  x = np.linspace(0.01, 15, 1000)
  x_indifference = np.linspace(0.05, z+borderlimit/2, 1000)
  plt.xlim(0,borderlimit)
  plt.ylim(0,borderlimit)
  plt.plot(x_indifference, indifference(x_indifference), label='indifference')
  plt.plot(x,contrainteBudgetaire(x), label='contrainte budgetaire')
  plt.title('choix sous contraintes')
  plt.xlabel('produit x')
  plt.ylabel('produit y')
  plt.grid(True)
  plt.legend()
  plt.show()
interact_manual(plot_graph,
         a=FloatText(value=-1.0, description='a < 0:'),
         b=FloatText(value=10, description='b:'),
         z=FloatText(value=1, description='z:'),
         c=FloatText(value=0.1, description='c:'));
```
